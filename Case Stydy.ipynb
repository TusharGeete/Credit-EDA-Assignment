```python
# Dividing the data into two dataset of Target1 = Defaulters |||| Target0 = Non Defaulters
Target0 = inp1[inp1.TARGET == 0]
Target1 = inp1[inp1.TARGET == 1]
inp1.TARGET.value_counts()
```

```python
#crating function to calculate percentage of 0's and 1's in "TARGET" column
def percentage_zeros(series):
    total_count = len(series)
    zero_count = (series == 0).sum()
    return (zero_count / total_count) * 100

def percentage_ones(series):
    total_count = len(series)
    one_count = (series == 1).sum()
    return (one_count / total_count) * 100
    
def count_zeroes(series):
    return (series == 0).sum()

def count_ones(series):
    return (series == 1).sum()

# Group by "CODE_GENDER" and plot count
agg_result = inp1.groupby("CODE_GENDER")["TARGET"].agg(count_zeroes=count_zeroes, count_ones=count_ones)
agg_result.plot.bar()
plt.ylabel('Count')
plt.title('Count of Zeros and Ones in TARGET by Gender')
plt.show()

# Group by "CODE_GENDER" and percentage
agg_result1 = inp1.groupby("CODE_GENDER")["TARGET"].agg(percentage_zeros=percentage_zeros, percentage_ones=percentage_ones)
agg_result1.plot.bar()
plt.ylabel('Percentage')
plt.title('Percentage of Zeros and Ones in TARGET by Gender')
plt.show()
```

Observation : 
1. The percentage of defaulters of Males is slightly higher than percentage of defaulters of Females
2. The number of Females applied for the loan is higher.

```python
# Group by "NAME_INCOME_TYPE" and plot count
agg_result = inp1.groupby("NAME_INCOME_TYPE")["TARGET"].agg(count_zeroes=count_zeroes, count_ones=count_ones)
agg_result.plot.bar()
plt.ylabel('Count')
plt.title('Count of Zeros and Ones in TARGET by Income Type')
plt.show()

# Group by "NAME_INCOME_TYPE" and percentage
agg_result1 = inp1.groupby("NAME_INCOME_TYPE")["TARGET"].agg(percentage_zeros=percentage_zeros, percentage_ones=percentage_ones)
agg_result1.plot.bar()
plt.ylabel('Percentage')
plt.title('Percentage of Zeros and Ones in TARGET by Income Type')
plt.show()
```
